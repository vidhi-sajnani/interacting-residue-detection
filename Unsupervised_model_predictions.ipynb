```python
from sklearn.cluster import KMeans
import pandas as pd
from matplotlib import pyplot as plt
```

```python
import pandas as pd

# Read the CSV file into a DataFrame
data = pd.read_csv("Dataset_for_unsupervied_model.csv")

# Select the columns you want to multiply by -1
columns_to_multiply = ['Leucine_Propensity', 'Threonine_Propensity', 'Methionine_Propensity', 'Serine_Propensity', 'Valine_Propensity']

# Multiply the selected columns by -1
data[columns_to_multiply] = data[columns_to_multiply].multiply(-1)

# Now, the specified columns have been multiplied by -1

aa_to_number = {
    'A': 1, 'C': 2, 'D': 3, 'E': 4, 'F': 5,
    'G': 6, 'H': 7, 'I': 8, 'K': 9, 'L': 10,
    'M': 11, 'N': 12, 'P': 13, 'Q': 14, 'R': 15,
    'S': 16, 'T': 17, 'V': 18, 'W': 19, 'Y': 20
}

# Convert Amino_Acid column to numbers
data['Amino_Acid'] = data['Amino_Acid'].map(aa_to_number)
data
```

```python
Sum_Of_Distances = []
K= range(1,10)
for k in K:
  km= KMeans(n_clusters=k)
  km= km.fit(data[['Rel-SASA','Amino_Acid','Methionine_Propensity','Serine_Propensity','Tryptophan_Propensity','Arginine_Propensity','Threonine_Propensity','Secondary_Enum','Leucine_Propensity','Tyrosine_Propensity','Valine_Propensity']])
  Sum_Of_Distances.append(km.inertia_)
```

```python
plt.plot(K,Sum_Of_Distances )
```

```python
data['Cluster']= Predicted_Clusters
data.head()
```

```python
output = 'output.csv'
data.to_csv(output, index=False)
```
